# NUMPY FUNCTIONS

```python
print("Mean Price - ", np.mean(df['product_price']))
print("Max Revenue - ", np.max(df['Revenue']))
print("Min Revenue - ", np.min(df['Revenue']))
print("Standard Deviation of Revenue - ", np.std(df['Revenue']))
```

# PANDAS ANALYSIS


## TOTAL REVENUE

```python
total_revenue = df['Revenue'].sum()
print("total_revenue - ",total_revenue)
```

## Category wise revenue

```python
Category_wise_revenue = df.groupby("category")['Revenue'].sum().sort_values(ascending = True)
print("Category_wise_revenue - ",Category_wise_revenue)
```

## City wise revenue

```python
City_wise_revenue = df.groupby("city")['Revenue'].sum().sort_values(ascending=True)
print("City_wise_revenue - ",City_wise_revenue)
```

## Customer_wise_revenue

```python
Customer_wise_revenue = df.groupby("customer_name")["Revenue"].sum().sort_values(ascending=True)
print("Customer_wise_revenue - ",Customer_wise_revenue)
```

```python
Month_wise_revenue = df.groupby('Month')['Revenue'].sum().sort_values(ascending=True)
print("Month_wise_revenue - ", Month_wise_revenue)
```

## Payment mode analysis

```python
payment_analysis = df.groupby('payment_mode')['Revenue'].sum()
payment_analysis
```

# CHARTS

## Correlation heatmap

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only = True)

plt.imshow(corr)
plt.xticks(range(len(corr.columns)),corr.columns, rotation = 90)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()
```

```python
## Revenue distribution
plt.figure()
plt.hist(df['Revenue'], label="Revenue",bins =10, edgecolor="black",alpha=0.5,rwidth=0.8,cumulative=True)
plt.title("Revenue distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.grid()
plt.legend()
plt.show()
```

## Product price distribution

```python
plt.figure()
plt.hist(df['product_price'],rwidth = 0.8, label="Product price", bins =10 ,edgecolor="black",cumulative = True)
plt.title('Product price distribution',fontweight = "bold")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.figure()
month = df["Month"]
revenue = df["Revenue"]
plt.bar(month,revenue,color = 'y', label="Product price", linewidth=1,edgecolor="black",width=0.5, align="center")
plt.title("Month wise Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

# BUSINESS QUESTION

### Q1  Which category gives highest revenue?

```python
category_revenue = df.groupby('category')["Revenue"].sum()
top_category = category_revenue.idxmax()
print("top_category = ",top_category)
```

### Q2  Which city performs best?

```python
city_revenue = df.groupby('city')['Revenue'].sum()
top_city = city_revenue.idxmax()
print("top_city - ",top_city)
```

### Q3  Does discount increase revenue?

```python
disc_revenue = df[['discount_percent',"Revenue"]]
discount_increase_revenue = disc_revenue.corr()
print(discount_increase_revenue)
print()
print("""        1. value close to +1  strong positive impact,
            2. value close to 0  no impact,
            3. value close to -1  negative impact""")
```

## Business conclusions

```python
category_revenue = df.groupby('category')["Revenue"].sum()
city_revenue = df.groupby('city')["Revenue"].sum()
customer_revenue = df.groupby('customer_name')["Revenue"].sum()
month_revenue = df.groupby('Month')["Revenue"].sum()
total_revenue = df["Revenue"].sum()

print("Top Category:", category_revenue.idxmax())
print("Top City:", city_revenue.idxmax())
print("Top Customer:", customer_revenue.idxmax())
print("Best Month:", month_revenue.idxmax())
print("Total Revenue:", total_revenue)
```
